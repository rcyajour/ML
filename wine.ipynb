```python
# -*- coding: utf-8 -*-
"""
Created on Tue Aug 27 22:26:45 2024

@author: rcyajour
"""

import pandas as pd
import csv

link = "wineQT.csv"
w = pd.read_csv(link)
w
```

```python
#storing all the column values as float.
# Stored as a list of dictionaries with each row as list containing dictionary of columns as key and data as values
wine=[]
w1=open('wineQT.csv', mode='r')
reader = csv.DictReader(w1)
for r in reader:   #r = row , k = key
    for k in r:
        r[k] = float(r[k])
       
    wine.append(r)
wine
```

```python
#QUALITY ATTRIBUTE VALUES :
# 0(bad) if x<=6 and 1(good) if x>6.
q='quality'
for r in wine:
    if r[q] > 6:
        r[q] = 1
    else:
        r[q] = 0
```

```python
wine[0].keys()
```

```python
#MIN MAX SCALING TECHNIQUE :

col =[]
for i in wine[0].keys():
   # print(i)
    if i!= q:       #Excluding 'quality' column from min max scaling
        col+=[i]
        
col

#creating two rows to store min and max values of each column
mins = {c: float('inf') for c in col}
maxs = {c: float('-inf') for c in col}

for r in wine:
    for c in col:
        if r[c] < mins[c]:
            mins[c] = r[c]
        if r[c] > maxs[c]:
            maxs[c] = r[c]
```

```python
mins,maxs
```

```python
# MIN-MAX SCALING
for r in wine:
    for c in col:
        r[c] = (r[c] - mins[c]) / (maxs[c] - mins[c])  #Normalization formula

```

```python
w2= open('wineQT_after_min_max.csv', mode='w', newline='')   #write normalized data to new file
writer = csv.DictWriter(w2, fieldnames=wine[0].keys())
writer.writeheader()
writer.writerows(wine)
```

## Why we use MIN- MAX SCALING?

Different features in the dataset wineQT have different units and scales (Ex: Density, pH, Acid Levels and Quality). 

Min-max scaling standardizes these features to a uniform scale of range 0 to 1 ensuring that no single feature dominates others due to its larger magnitude.

This makes interpretation easier and relatively simpler making algorithms that rely on distance calculations, like k-means clustering and k-nearest neighbors, perform better.
